## 1. Gausssian kernel 그리기

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def GaussianKernel(v1, v2, sigma):
    temp= np.power(v1-v2, 2)
    temp= np.sum(temp)/(2*sigma)
    res= np.exp(-temp)
    return res
```

```python
val= GaussianKernel(np.array([0,0]), np.array([3,5]), 1.0)
print val
```

```python
d=100
x1= np.linspace(0, 6, d) # 0~6 (1차원)
x2= np.linspace(0, 10, d) # 0~10 (1차원)
l1=[3,5]

#sigma-=0.5
Z_05= np.zeros((d,d))
sigma= 0.5
for i, t1 in enumerate(x1):
    for j, t2 in enumerate(x2):
        val_05= GaussianKernel(np.array([t1, t2]), l1, sigma)
        Z_05[i, j]= val_05
    
#sigma= 1.0
Z_10= np.zeros((d,d))
sigma= 1.0
for i, t1 in enumerate(x1):
    for j, t2 in enumerate(x2):
        val_10= GaussianKernel(np.array([t1, t2]), l1, sigma)
        Z_10[i, j]= val_10
        
#sigma= 1.5
Z_15= np.zeros((d,d))
sigma= 1.5
for i, t1 in enumerate(x1):
    for j, t2 in enumerate(x2):
        val_15= GaussianKernel(np.array([t1, t2]), l1, sigma)
        Z_15[i, j]= val_15
```

```python
from mpl_toolkits import mplot3d
pX, pY= np.meshgrid(x1, x2)

fig= plt.figure(figsize=(40,10))
# ax= plt.axes(projection= '3d')

#sigma-=0.5
ax= fig.add_subplot(1,3,1, projection='3d')
ax.contour3D(pX, pY, Z_05, 100, cmap='jet')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_ylabel('Z')
plt.title('sigma= 0.5', fontsize=25)

#sigma-=1.0
ax= fig.add_subplot(1,3,2, projection='3d')
ax.contour3D(pX, pY, Z_10, 100, cmap='jet')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_ylabel('Z')
plt.title('sigma= 1.0', fontsize=25)

#sigma-=1.5
ax= fig.add_subplot(1,3,3, projection='3d')
ax.contour3D(pX, pY, Z_15, 100, cmap='jet')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_ylabel('Z')
plt.title('sigma= 1.5', fontsize=25)

plt.show()
```

## 선형 SVM 분류

### 1) 데이터 읽기

```python
import numpy as np
import os
from scipy.io import loadmat

data= loadmat(os.path.join('data', 'data1.mat'))
print data

X= data['X'] #x1, x2의 값
y= data['y'] # 0 or 1
m= len(y)
```

```python
print X.shape, y.shape
```

### 2) 그래프 그리기

```python
pos= []
neg= []

for i, val in enumerate(y):
    if val==1:
        pos.append(i)
    else:
        neg.append(i)
print pos
print neg
```

```python
import matplotlib.pyplot as plt
#X[:, 1].reshape(-1) -> 한 줄로 피기 (m,) -> (m)
plt.plot(X[pos, 0].reshape(-1), X[pos, 1].reshape(-1), 'b+', label='y=1')
plt.plot(X[neg, 0].reshape(-1), X[neg, 1].reshape(-1), 'ro', label='y=0')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper right')
plt.title("Data 1")
plt.show()
```

### 3) 학습

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
# 평균0, 표준편차1로 만듦 (feature normalization)
from sklearn.svm import LinearSVC

svm_clf_c1= Pipeline([
    ("scalar", StandardScaler()),
    ("linear_svc", LinearSVC(C=1, loss='hinge'))
])

svm_clf_c100= Pipeline([
    ("scalar", StandardScaler()),
    ("linear_svc", LinearSVC(C=100, loss='hinge'))
])

svm_clf_c1.fit(X, y)
svm_clf_c100.fit(X, y)
```

```python
result_c1= svm_clf_c1.predict([[2.3, 2.7],
                              [3.7, 3.9]])
result_c100= svm_clf_c100.predict([[2.3, 2.7],
                              [3.7, 3.9]])
print(result_c1)
print(result_c100)
```

### 4) Decision boundary

```python
x_min, x_max= X[:, 0].min()-0.2, X[:, 0].max()+0.2
y_min, y_max= X[:, 1].min()-0.2, X[:, 1].max()+0.2
h= .02 #step size in the mesh
xx, yy= np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
plt.figure(figsize=(15, 5))

# C= 1
plt.subplot(1, 2, 1)
Z_c1= svm_clf_c1.predict(np.c_[xx.ravel(), yy.ravel()])
Z_c1= Z_c1.reshape(xx.shape)
plt.figure(1)
plt.pcolormesh(xx, yy, Z_c1, cmap=plt.cm.Paired)

plt.plot(X[pos, 0].reshape(-1), X[pos, 1].reshape(-1), 'b+', label='y=1')
plt.plot(X[neg, 0].reshape(-1), X[neg, 1].reshape(-1), 'ro', label='y=0')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper right')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())
plt.title("C= 1")

# C= 100
plt.subplot(1, 2, 2)
Z_c100= svm_clf_c100.predict(np.c_[xx.ravel(), yy.ravel()])
Z_c100= Z_c100.reshape(xx.shape)
plt.figure(1)
plt.pcolormesh(xx, yy, Z_c100, cmap=plt.cm.Paired)

plt.plot(X[pos, 0].reshape(-1), X[pos, 1].reshape(-1), 'b+', label='y=1')
plt.plot(X[neg, 0].reshape(-1), X[neg, 1].reshape(-1), 'ro', label='y=0')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper right')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())
plt.title("C= 100")


plt.show()
```

## 3. 비선형 SVM 분류

### 1) 데이터 읽기

```python
import numpy as np
import os
from scipy.io import loadmat

data= loadmat(os.path.join('data', 'data2.mat'))
print data

X= data['X'] #x1, x2의 값
y= data['y'] # 0 or 1
m= len(y)
```

```python
print X.shape, y.shape
```

### 2) 그래프 그리기

```python
pos= []
neg= []

for i, val in enumerate(y):
    if val==1:
        pos.append(i)
    else:
        neg.append(i)
print pos
print neg
```

```python
import matplotlib.pyplot as plt
#X[:, 1].reshape(-1) -> 한 줄로 피기 (m,) -> (m)
plt.plot(X[pos, 0].reshape(-1), X[pos, 1].reshape(-1), 'b+', label='y=1')
plt.plot(X[neg, 0].reshape(-1), X[neg, 1].reshape(-1), 'ro', label='y=0')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper right')
plt.title("Data 2")
plt.show()
```

### 3) 학습

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
# 평균0, 표준편차1로 만듦 (feature normalization)
from sklearn.svm import SVC

svm_clf= Pipeline([
    ("scalar", StandardScaler()),
    ("svc", SVC(kernel="rbf", gamma=10, C=1))
])

svm_clf.fit(X, y)
```

### 4) Decision boundary

```python
x_min, x_max= X[:, 0].min()-0.2, X[:, 0].max()+0.2
y_min, y_max= X[:, 1].min()-0.2, X[:, 1].max()+0.2
h= .01 #step size in the mesh
xx, yy= np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
# plt.figure(figsize=(15, 5))

Z= svm_clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z= Z.reshape(xx.shape)
plt.figure(1)
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

plt.plot(X[pos, 0].reshape(-1), X[pos, 1].reshape(-1), 'b+', label='y=1')
plt.plot(X[neg, 0].reshape(-1), X[neg, 1].reshape(-1), 'ro', label='y=0')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper right')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())
plt.title("C= 1")

plt.show()
```
